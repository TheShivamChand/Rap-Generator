```python
# Import necessary libraries

import numpy as np
from gtts import gTTS
```

```python
# function to load the text file in the notebook

def load_text(filename):
    with open(filename,encoding="utf-8") as f:
        return f.read().lower()
```

```python
# load the text file

data=load_text("lyrics.txt")
```

```python
# function to make a dictionary of words of size 4 containing the words and the corresponding upcoming
# character with its frequency

def transactionTable(data,k=4):
    T={}
    for i in range(len(data)-k):
        X=data[i:i+k]
        Y=data[i+k]
        if T.get(X) is None:
            T[X]={}
            T[X][Y]=1
        elif T[X].get(Y) is None:
            T[X][Y]=1
        else:
            T[X][Y]+=1
    return T
```

```python
test= transactionTable(data)
```

```python
# for each key in the dictionary, we make a denominator_factor which will be used to make the probabililty
# of every character which will be coming next in the word

def probability(T):
    for kx in T.keys():
        denom_factor = sum(T[kx].values())
        for val in T[kx].keys():
            T[kx][val]/=denom_factor
    return T
```

```python
test_pro=probability(test)
```

```python
test_pro
```

```python
#here we take a string from the data and make two lists. The first list will contain the word and
# the second one will contain the probability of the word

def predict(string, T):
    if T.get(string) is None:
        return " "
    next_char=list(T[string].keys())
    probab = list(T[string].values())
    return np.random.choice(next_char, p=probab)
```

```python
# we generate a senctence with a word and the remaining characters will be generated
# automatically as per their probability 

sentence = " कौन"
for i in range(1000):
    next_char = predict(sentence[-4:], test_pro)
    sentence+=next_char
print(sentence)
```

```python
# we convert the senctence into a new file and save it in mp3 format using gTTS

file = gTTS(sentence, lang="hi")
file.save("rap.mp3")
```
